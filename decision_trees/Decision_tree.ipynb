```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from IPython.display import Image
```

```python
iris = load_iris()
X = iris.data[:, 2:]
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth = 2)
tree_clf.fit(X, y)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(tree_clf, out_file = 'iris_tree.dot', feature_names = iris.feature_names[2:],
               class_names = iris.target_names, rounded = True, filled = True)
```

```python
image = 'iris_tree.png'
Image(image)
```

## Fazendo predicao

```python
tree_clf.predict_proba([[5, 1.5]])
```

## Regularization Hyperparameters

Para evitar overfit nos dados de treino e necessario restringir a liberdade da decision tree durante o treinamento<br>
Isso e ja e conhecido como regularizacao como mostrado em regressao linear diminuindo o max_depth ja e um comeco<br>
Outros parametros para alterar o formato da arvore e min_samples_split, min_samples_leaf, min_weight_fraction_leaf<br>
max_leaf_nodes, max_features. Aumentando o min_* e reduzindo o max_* hiperparametros irao regularizar o modelo

# Regressao

```python
from sklearn.tree import DecisionTreeRegressor
```
