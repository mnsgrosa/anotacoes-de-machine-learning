## Manifold Learning

Uma coleção de pontos formando um conjunto que aquela superficie fechada é analoga a uma dimensão superior. Isso é a definição de um manifold. A dimensão do manifold é d e é sempre menor que sua forma original n ou seja d < n. A hipotese de manifold é sempre acompanhado por outra assunção e ela é: A tarefa (classificação ou regressão) será mais simples se expressa em uma dimensionalidade mais baixa do manifold. Por exemplo no swiss roll em 3D teria 2 classes e no 2D teria apenas uma linha dividindo o dataset. Mas nem sempre é real. Manifold Learning nem sempre simplifica o processo. Always remember that

## PCA (principal component analysis)

É o algoritmo mais popular de redução de dimensionalidade. Primeiro ele identifica o hyperplano que está mais proximo do plano e projeta o dado nele assim como na figura 8.2 acima nesse notebook.

### Preservando a variancia

Antes de projetar os dados tem de escolher o hyperplano certo e o hyperplano certo é aquele que tem mais variância por 2 motivos: Um deles é que maior variancia quer dizer mais informação e por fim ele minimiza a distância média quadrada entre o dataset original e sua projeção no eixo.
